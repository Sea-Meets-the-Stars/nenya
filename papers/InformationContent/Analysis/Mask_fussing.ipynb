# Mask tests

```python
# imports
from importlib import reload
import numpy as np

from wrangler.plotting import cutout
from wrangler import utils as wr_utils
```

# Fake iamge

```python
img = np.random.uniform(size=(20,20))
```

```python
cutout.show_image(img)
```

# Mask 

```python
mask = np.zeros_like(img, dtype=bool)
mask[10,10] = True
```

## Check

```python
def show_masked(iimg, mask):
    mimg = iimg.copy()
    mimg[mask] = -100.
    cutout.show_image(mimg)
```

```python
show_masked(img, mask)
```

# Grow me

## 1.5

```python
grow_msk = wr_utils.grow_mask(mask, 1.5)
```

```python
show_masked(img, grow_msk)
```

## 1.25

```python
grow_msk = wr_utils.grow_mask(mask, 1.25)
show_masked(img, grow_msk)
```
